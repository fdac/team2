```python
import requests, json, sys, time, re, BeautifulSoup, time

'''
def process_commits(url):
    print url
    r = requests.get(url)
    t = r.text
    data = json.loads(t)
    
    try:
        data['error']
        return
    except KeyError:
        pass
    
    for commit in data['values']:
        #print commit['author']['user']['username']
        pass
    
    try:
        print data['next']
        #process_commits(data['next'])
    except KeyError:
        return
'''

# grab a url (contains roughly 10 repos) and process it
def process_repos(url, start_time):
    python_count = 0
    set_count = 0
    total_count = 0
    
    # get the url data (a json file) and put it in a dictionary
    r = requests.get(url)
    t = r.text
    data = json.loads(t)
    
    while True:
        # do something for each repo
        for repo in data['values']:
            total_count += 1
            
            # 2**29 ~ 500 MB
            if repo['size'] > 2**29:
                continue
                
            if not repo['language'] == '':
                set_count += 1
            
            if not repo['language'] == 'python':
                continue
            
            python_count += 1
        
        # check the time elapsed, return after a bit of time
        if (time.time() - start_time) > 6*60*60:
            return (python_count, set_count, total_count)
        
        # grab the next page of repos (if it exists, done otherwise)
        try:
            r = requests.get(data['next'])
            t = r.text
            data = json.loads(t)
        except KeyError:
            return (python_count, set_count, total_count)

# driver for the example
start_time = time.time()
URL = 'https://bitbucket.org/api/2.0/repositories/'
py_count, set_count, total_count = process_repos(URL, start_time)
print 'total # of target repos: ' + str(py_count)
print 'total # of repos analyzed: ' + str(set_count)
print 'pct: ' + str(1.0*py_count / set_count)
print 'total # of public repos: ' + str(total_count)
```
